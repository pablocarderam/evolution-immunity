```python
def plotdist(dist):
     sns.distplot(dist, hist=False)
```

```python
mu, sigma = 0.5, 0.1
normal_dist = np.random.normal(mu, sigma, 1000)
plotdist(normal_dist/normal_dist.sum())
```

```python
w1 = [0.4,0.25,0.15,0.1,0.1]  # decaying weights
w2 = [0.1,0.1,0.15,0.25,0.4]  # increasing weights
w3 = [0.1,0.2,0.4,0.2,0.1]    # bell weights
pmw1 = [1]      # Perfect matching weight
pmw2 = [0.5]    # Perfect matching weight 

exps = {
    'decaying' : w1,
    'increasing' : w2,
    'bell' : w3,
    'perfect_matching_all' : pmw1,
    'perfect_matching_half' : pmw2,
}
```

```python
from scipy import interpolate


for k, i in exps.items():
    if len(i) == 1:
        i = np.repeat(i, 5)
    df = pd.DataFrame()
    df['pos'] = [0,1,2,3,4]
    df['weight'] = i
    fig, ax = plt.subplots(1, 1, figsize=(5,5))
    sns.barplot(ax=ax, data=df,x='pos', y='weight', label=k)
    #eliminate x axis
    ax.set_xlabel('')
    ax.set_ylim(0,1.01)
    # ax.set_xticks([])
    ax.set_ylabel('Weight')
```
